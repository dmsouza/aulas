Importando as bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [10, 6]
%config InlineBackend.figure_format = 'retina'
```

Conectando a um site e coletando informações

```python
df = pd.read_html('https://www.ibge.gov.br/explica/inflacao.php')[1]
```

Monstrando os dados coletados

```python
df.head(50)
```

Renomeando as colunas

```python
df = df.rename({
    'IPCA [Fev/2022]': 'ipca',
    'INPC [Fev/2022]': 'inpc'
}, axis = 1)
```

Formatando os dados

```python
df['ipca'] = df['ipca'].str.replace('%','')
df['inpc'] = df['inpc'].str.replace('%','')

df['ipca'] = df['ipca'].str.replace(',','.')
df['inpc'] = df['inpc'].str.replace(',','.')
```

Convertendo os dados para os tipos correspondentes

```python
df['ipca'] = df['ipca'].astype(float)
df['inpc'] = df['inpc'].astype(float)
```

```python
df.plot(kind='bar', x='Local')

plt.xticks(horizontalalignment="center")

plt.title("IPCA - Estados (Fevereiro/22)")
plt.xlabel("Variação por Estado")
plt.ylabel("Variação")

plt.show()
```
